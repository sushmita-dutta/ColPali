```python
import os
import base64
from io import BytesIO
from PIL import Image
import gradio as gr
from byaldi import RAGMultiModalModel

# Set your HuggingFace token
os.environ["HF_TOKEN"] = ""

# Load ColPali model once
RAG = RAGMultiModalModel.from_pretrained("vidore/colpali-v1.2", verbose=1)

# Define the function that takes PDF + question and returns image + text
def search_pdf(pdf_file, query):
    try:
        # Save PDF
        with open("temp_input.pdf", "wb") as f:
            f.write(pdf_file)

        # Index the file
        RAG.index(
            input_path="temp_input.pdf",
            index_name="attention",
            store_collection_with_index=True,
            overwrite=True
        )

        # Search the indexed PDF
        results = RAG.search(query, k=1)

        if not results:
            return None, "No results found."

        # Decode the image
        image_bytes = base64.b64decode(results[0].base64)
        image = Image.open(BytesIO(image_bytes))

        # ✅ FIXED: Extract matched text
        matched_text = results[0].metadata.get("text", "No text found.")

        return image, matched_text

    except Exception as e:
        return None, f"Error: {str(e)}"



# Create Gradio interface
with gr.Blocks() as demo:
    gr.Markdown("## 🧠 Visual RAG with ColPali")
    gr.Markdown("Upload a PDF and ask a question. The model returns the most relevant page and matching text.")

    pdf_input = gr.File(label="📄 Upload PDF", type="binary")
    query_input = gr.Textbox(label="❓ Ask a Question", placeholder="e.g., Where does intrafusal muscle fibers lie?")
    submit_button = gr.Button("🔍 Search")

    result_image = gr.Image(label="📸 Most Relevant Page")
    result_text = gr.Textbox(label="📄 Matched Text")


    submit_button.click(fn=search_pdf, inputs=[pdf_input, query_input], outputs=[result_image, result_text])

demo.launch(share=True)
```
